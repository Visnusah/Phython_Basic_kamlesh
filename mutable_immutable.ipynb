## Mutable and Immutable

```python
username = "KamleshSah"
username = "KamleshKumarSah"
print(username) # KamleshKumarSah
```

```python
x = 10
y = x
print(x) # 10
print(y) # 10
x = 15 # x will change but y will not change
print(y) # 10 , y will not change because it is not a reference to x
print(x) # 15 , x will change because it is a reference to x
```
